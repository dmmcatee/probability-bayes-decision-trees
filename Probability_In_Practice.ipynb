**Simulate A Binomial Experiment**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_flips = 10
p_heads = 0.5
n_trials = 1000

heads_counts = np.random.binomial(n=n_flips, p=p_heads, size=n_trials)

heads_counts[:10]
```

```python
plt.figure(figsize=(8,5))
plt.hist(heads_counts, bins=np.arange(-0.5, n_flips + 1.5, 1), edgecolor = "black")
plt.xticks(range(0, n_flips + 1))
plt.xlabel("Number of Heads (Out of 10)")
plt.ylabel("Frequency (Out of 1000 Trials)")
plt.title("Distribution of Heads in 10 Coin Flips (1000 Trials)")
plt.show()
```

```python
empirical_prob_5 = np.mean(heads_counts == 5)
empirical_prob_5
```

```python
print(f"Empirical P(exactly 5 heads) = {empirical_prob_5: .4f}")
```

**Conditional Probability with Data**

```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
df = titanic[["sex", "survived"]]
df["sex"].value_counts(), df["survived"].value_counts()
```

```python
p_survived_given_female = df[df["sex"] == "female"]["survived"].mean()
print(f"p(survived | female) = {p_survived_given_female: .4f}")
```

```python
survivors = df[df["survived"] == 1]
p_female_given_survived = (survivors["sex"] == "female").mean()

print(f"p(female | survived) = {p_female_given_survived:.4f}")
```

```python
pd.crosstab(df["sex"], df["survived"], margins = True)
```
